```python
for i in range(1, 429):
    url = f"https://api.themoviedb.org/3/movie/top_rated?language=en-US&page={i}"
    headers = {
        "accept": "application/json",
        "Authorization": "Bearer eyJhbGciOiJIUzI1NiJ9.eyJhdWQiOiIxZDAzZWZhMTg0ODEwNzQ3NmZlOWUwZmQ5NzU0ZDg4ZSIsIm5iZiI6MTc0NTE2NTExNS45NDcsInN1YiI6IjY4MDUxYjNiYWMwMmQ0NDA3YmFhOTQ3ZSIsInNjb3BlcyI6WyJhcGlfcmVhZCJdLCJ2ZXJzaW9uIjoxfQ.Gp1-YD8RjHM50clDD_pf-ynfepLHZa2Z0xhiZ79p9EA"
    }

    try:
        response = requests.get(url, headers=headers)
        response.raise_for_status()  # Raise exception for HTTP errors
        results = response.json().get('results', [])
        temp_df = pd.DataFrame(results)[['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']]
        df = pd.concat([df, temp_df], ignore_index=True)
    except Exception as e:
        print(f"Error on page {i}: {e}")
```
