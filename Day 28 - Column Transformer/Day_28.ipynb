# Column Transformer

## What is a Column Transformer?

A Column Transformer is a preprocessing tool in scikit-learn that allows us to apply different preprocessing techniques to different columns of a dataset at the same time.

For example:
- We can apply One-Hot Encoding to categorical columns.
- We can apply Standard Scaling to numerical columns.
- We can leave the remaining columns unchanged.

Instead of preprocessing each column separately, a Column Transformer combines everything into a single step.

---

## Why do we use a Column Transformer?

In real-world datasets, not all columns are of the same type.

For example:

- Age → Numerical
- Salary → Numerical
- Gender → Categorical
- City → Categorical

Since different types of data require different preprocessing techniques, we use a Column Transformer to apply the correct transformation to each type of column.

---

## Syntax

```python
from sklearn.compose import ColumnTransformer

ct = ColumnTransformer(
    transformers=[
        ('encoder', OneHotEncoder(drop='first'), [0]),
        ('scaler', StandardScaler(), [1,2])
    ],
    remainder='passthrough'
)
```

---

## Explanation of the Parameters

### transformers

This parameter contains a list of transformations.

Each transformation has three parts:

```python
('name', transformer, columns)
```

Example:

```python
('encoder', OneHotEncoder(), [0])
```

- 'encoder' → Name of the transformation.
- OneHotEncoder() → Transformation to apply.
- [0] → Column index on which the transformation will be applied.

---

### remainder='passthrough'

This tells the Column Transformer what to do with the columns that are not mentioned in the `transformers` list.

There are two options:

- `passthrough` → Keep the remaining columns unchanged.
- `drop` → Remove the remaining columns.

Most of the time, we use:

```python
remainder='passthrough'
```

---

## Advantages of Column Transformer

- Applies different preprocessing methods to different columns.
- Reduces code complexity.
- Prevents manual preprocessing mistakes.
- Makes the preprocessing pipeline cleaner and easier to manage.
- Useful for real-world datasets with both numerical and categorical features.

## Interview Questions

### Q1. What is a Column Transformer?

Answer:
A Column Transformer is a scikit-learn tool that allows us to apply different preprocessing techniques to different columns of a dataset in a single step.

---

### Q2. Why do we use a Column Transformer?

Answer:
We use it because numerical and categorical columns require different preprocessing methods. It applies the appropriate transformation to each column automatically.

---

### Q3. What does `remainder='passthrough'` mean?

Answer:
It keeps all the columns that are not mentioned in the transformers list unchanged.

---

### Q4. What does `remainder='drop'` mean?

Answer:
It removes all the columns that are not specified in the transformers list.

---

### Q5. Can we apply multiple transformations using a Column Transformer?

Answer:
Yes. We can apply different transformations like One-Hot Encoding, Standard Scaling, Min-Max Scaling, or Ordinal Encoding to different columns at the same time.

## Simple Example

Suppose we have the following dataset:

| Age | Salary | Gender | City |
|-----|---------|---------|------|
|25|30000|Male|Mumbai|
|30|50000|Female|Pune|

We want to:
- One-Hot Encode → Gender and City
- Standard Scale → Age and Salary

A Column Transformer allows us to do all of this together instead of writing separate preprocessing code for each column.

````python
import pandas as pd
import numpy as np
````

````python

from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
from sklearn.preprocessing import OrdinalEncoder
````

````python
df = pd.read_csv("covid_toy.csv")
````

````python
df.head()
````

````python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(df.drop(columns=['has_covid']),df['has_covid'],
                                                test_size=0.2)
````

````python
from sklearn.compose import ColumnTransformer
````

````python
transformer = ColumnTransformer(transformers=[
    ('tnf1',SimpleImputer(),['fever']),
    ('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('tnf3',OneHotEncoder(sparse_output=False,drop='first'),['gender','city'])
],remainder='passthrough')
````

````python
transformer.fit_transform(X_train).shape
````

````python

transformer.transform(X_test).shape
````
